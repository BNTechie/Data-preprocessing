```python
#import sys
#import os
#!{sys.executable} -m pip install numpy==1.23.5
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler, Normalizer
```

```python


# Load the iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Apply different scalers
scalers = {
    "Min-Max Scaling": MinMaxScaler(),
    "Standardization": StandardScaler(),
    "Robust Scaler": RobustScaler(),
    "MaxAbs Scaler": MaxAbsScaler(),
    "L2 Normalization": Normalizer(norm='l2')
}

scaled_data = {}
```

```python

for key, scaler in scalers.items():
    scaled_data[key] = scaler.fit_transform(data)

# Plotting the original and scaled data
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 20))
axes = axes.flatten()

# Original data scatter plot
axes[0].scatter(data.iloc[:, 0], data.iloc[:, 1], c=iris.target, cmap='viridis', edgecolor='k')
axes[0].set_title("Original Data")
axes[0].set_xlabel(iris.feature_names[0])
axes[0].set_ylabel(iris.feature_names[1])

# Plot each normalized dataset
for ax, (key, scaled) in zip(axes[1:], scaled_data.items()):
    ax.scatter(scaled[:, 0], scaled[:, 1], c=iris.target, cmap='viridis', edgecolor='k')
    ax.set_title(key)
    ax.set_xlabel(iris.feature_names[0])
    ax.set_ylabel(iris.feature_names[1])

# Remove empty subplot
fig.delaxes(axes[5])

plt.tight_layout()
plt.show()
```
